```python
# 'price' ustunini ajratib olamiz va X va y'ni yaratamiz
X = df_train.drop('price', axis=1)
y = df_train.price  # 'price' ustuni y bo'ladi

# X dagi kategorik ustunlarni One-Hot Encoding raqamli qiymatlarga o'zgartiramiz
X = pd.get_dummies(X)

# X ni chiqaramiz
X 
```

```python
# Ma'lumotlarni o'qitish va test to'plamlariga bo'lish uchun kutubxonani import qilamiz
from sklearn.model_selection import train_test_split  

# X va y ma'lumotlarini train va test to'plamlariga bo'lamiz
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True, random_state=73)
# X_train ma'lumotlarida har bir ustundagi noyob qiymatlar sonini hisoblaymiz
X_train.nunique()
```

```python
# MinMaxScaler obyektini yaratamiz
scr = MinMaxScaler()

# X_train ma'lumotlarini normalizatsiya qilamiz (0 va 1 orasidagi qiymatlarga o'zgartiramiz)
X_train_pr = scr.fit_transform(X_train)

# X_test ma'lumotlarini normalizatsiya qilamiz (faqat o'rgatish parametrlarini X_train'dan olib, test to'plamiga qo'llaymiz)
X_test_pr = scr.transform(X_test)

# X_train_prepared'ni ko'rsatamiz
X_train_pr
```

**DecisionTreeRegressor** modelini yaratamiz va fit metodi yordamida X_train_pr va y_train ma'lumotlari bilan o'qitamiz.
predict metodi yordamida test to'plamida bashorat qilamiz (X_test_pr).
Modelning xatoliklarini, ya'ni MAE (Mean Absolute Error) va RMSE (Root Mean Squared Error) hisoblaymiz va ekranda chiqaramiz👇👇👇.




```python
# DecisionTreeRegressor modelini yaratamiz
m_tree = DecisionTreeRegressor()

# Modelni o'qitish (train) to'plami bilan moslashtiramiz
m_tree.fit(X=X_train_pr, y=y_train)

# Test to'plami bilan bashorat qilamiz
prediction_rf = m_tree.predict(X_test_pr)

# Modelning xatoliklarini hisoblaymiz
mae = mean_absolute_error(y_test, prediction_rf)
mse = mean_squared_error(y_test, prediction_rf)

# Natijalarni ekranga chiqaramiz
print("Mean Absolute Error=", mae)
print("Root Mean Squared Error=", np.sqrt(mse))
```

**RandomForestRegressor**

```python
# RandomForestRegressor modelini yaratamiz
m = RandomForestRegressor(n_estimators=200, random_state=7575)

# Modelni o'qitish (train) to'plami bilan moslashtiramiz
m.fit(X_train_pr, y_train)

# Test to'plami bilan bashorat qilamiz
predict_test = m.predict(X_test_pr)

# Modelning xatoliklarini hisoblaymiz
mae = mean_absolute_error(y_test, predict_test)
mse = mean_squared_error(y_test, predict_test)

# Natijalarni ekranga chiqaramiz
print("mean_absolute_error=", mae)
print("root_mean_squared_error=", np.sqrt(mse))
```

**XGBRegressor**

```python
# XGBRegressor modelini yaratamiz
m_xgb = XGBRegressor()

# Modelni o'qitish (train) to'plami bilan moslashtiramiz
m_xgb.fit(X=X_train_pr, y=y_train)

# Test to'plami bilan bashorat qilamiz
prediction_rf = m_xgb.predict(X_test_pr)

# Modelning xatoliklarini hisoblaymiz
mae = mean_absolute_error(y_test, prediction_rf)
mse = mean_squared_error(y_test, prediction_rf)

# Natijalarni ekranga chiqaramiz
print("mean_absolute_error=", mae)
print("root_mean_squared_error=", np.sqrt(mse))
```
